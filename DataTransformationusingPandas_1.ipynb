```python
import pandas as pd
```

```python
#creating the datasheet
data1 ={'key':['k0','k1','k2','k3'],
       'Name':['Jai','Princi','Gaurav','Anuj'],
       'Age':[23,24,25,26]}
data2={'key':['k0','k1','k2','k3'],
      'Address':['Nagpur','Kanpur','Allahbad','Aurangabad'],
      'Qualification':['BTech','BCA','BE','BBA']}
#converting the dictionaries to dataframes
df=pd.DataFrame(data1)
df1=pd.DataFrame(data2)
print(df,'\n\n',df1)
```

```python
#merge with a single key
res = pd.merge(df,df1,on='key')
print(res)
```

```python
#creating the datasheet
data1 ={'Key':['k0','k1','k2','k3'],
        'Key1':['k0','k1','k2','k3'],
       'Name':['Jai','Princi','Gaurav','Anuj'],
       'Age':[23,24,25,26]}
data2={'Key':['k0','k1','k2','k3'],
       'Key1':['k0','k1','k2','k3'],
      'Address':['Nagpur','Kanpur','Allahbad','Aurangabad'],
      'Qualification':['BTech','BCA','BE','BBA']}
#converting the dictionaries to dataframes
df=pd.DataFrame(data1)
df1=pd.DataFrame(data2)
print(df,'\n\n',df1)
```

```python
#merge with multiple key (composite key)
res=pd.merge(df,df1,on=['Key','Key1'])
print(res)
```

```python
#creating the datasheet
data1 ={'Key':['k0','k1','k2','k3'],
        'Key1':['k0','k1','k0','k1'],
       'Name':['Jai','Princi','Gaurav','Anuj'],
       'Age':[23,24,25,26]}
data2={'Key':['k0','k1','k2','k3'],
       'Key1':['k0','k0','k0','k0'],
      'Address':['Nagpur','Kanpur','Allahbad','Aurangabad'],
      'Qualification':['BTech','BCA','BE','BBA']}
#converting the dictionaries to dataframes
df=pd.DataFrame(data1)
df1=pd.DataFrame(data2)
print(df,'\n\n',df1)
```

```python
#merge with multiple column (composite key) using the how argument
res=pd.merge(df,df1,how='left',on=['Key','Key1'])
print(res)        #all values from left df will be there
```

```python
#merge with multiple column (composite key) using the how argument
res=pd.merge(df,df1,how='right',on=['Key','Key1'])
print(res)            #all values from left df will be there
```

```python
#merge with multiple column (composite key) using the how argument
res=pd.merge(df,df1,how='outer',on=['Key','Key1'])
print(res)            #merge all the different rows
```

```python
#merge with multiple column (composite key) using the how argument
res=pd.merge(df,df1,how='inner',on=['Key','Key1'])
print(res)            
```

```python
#merge with multiple column (composite key) using the how argument
res=pd.merge(df,df1,on=['Key','Key1'])
print(res)            
```

```python
#creating the datasheet
data1 ={'Name':['Jai','Princi','Gaurav','Anuj'],
       'Age':[23,24,25,26],
       'Key':['K0','K1','K2','K3']}
data2={'Address':['Nagpur','Kanpur','Allahbad','Aurangabad'],
      'Qualification':['BTech','BCA','BE','BBA']}
#converting the dictionaries to dataframes
df=pd.DataFrame(data1)
df1=pd.DataFrame(data2,index=['K0','K2','K3','K4'])
print(df,'\n\n',df1)
```

```python
#JOIN OPERATION
res=df.join(df1,on='Key')
print(res)
```

```python
#defining a dictionary containing employess data
data1={'Name':['Jai','Princi','Gaurav'],
      'Age':[27,24,22]}
data2 ={'Address':['Allahbad','Kannauj','Allahbad','Kanpur'],
       'Qualification':['MCA','PhD','Bcom','B.Hons']}
#converting dictionaries to dataframe
df=pd.DataFrame(data1,index=pd.Index(['K0','K1','K2'],name='Key'))
index=pd.MultiIndex.from_tuples([('K0','Y0'),('K1','Y1'),
                                 ('K2','Y2'),('K2','Y3')],
                                names=['Key','Y'])
df1=pd.DataFrame(data2,index=index)
print(df,'\n\n',df1)
```

```python
#joining single index with multi index
res=df.join(df1,how='inner')
print(res)
```
