```python
# importing important libraries
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image,ImageOps
```

```python
#to open an image , open() function in Image modue of PIL library is useful.
#similarly matplotlib uses 2 methods-
#imread():to read the image
#imshow(: to display the image
#PIL : python Imaging Library)
```

```python
#importing the image data
img =np.array(Image.open('Tiger_shutterstock.jpeg'))
plt.figure(figsize=(8, 8))
plt.imshow(img)
```

```python
#getting the details of the image
print('Number of Dimension :',img.ndim)
print('image shape :',img.shape)
print('dtype  :',img.dtype)
print(img[20, 20]) # this will give pixel value for RGC vector
print(img[:, :,2].min()) # blue color vector min intensity
print(img[:, :,1].min())# green color vector min intensity
print(img[:, :,0].min()) # red color vector min intensity
print(img[:, :,0].max())   # red color vector max intensity
```

```python
#savinf an Image from n-dimenstional vector
path = 'Tiger.jpeg'
pil_img = Image.fromarray(img)
pil_img.save(path)
```

```python
#creating 1-D view
img= Image.open('Tiger_shutterstock.jpeg')
arr = np.array(img)

#recording shape of original array 
shape =arr.shape
print('shape  :',shape)

#making it i dimensional 
flat_arr = arr.ravel()
print('flattened Array :',flat_arr)


#converting it back to a matrix
vector = np.matrix(flat_arr)
print('Multi Dimensional Vector :',vector)
```

```python
#making some changes
vector[:, ::10]=120

#reform a numpy array of original shape 
arr2= np.asarray(vector).reshape(shape)

#making a PIL Image
img2 = Image.fromarray(arr2,'RGB')
img2.show()
```

```python
#histogram for pixel densities
img = np.array(Image.open('Tiger_shutterstock.jpeg'))
img_flat = img.flatten()
plt.hist(img_flat, bins=200, range=[0,256])
plt.title('Number of pixel in each intensity value')
plt.xlabel('Intensity')
plt.ylabel('NUmber of Pixel')
plt.show()
```

```python
# the above plot will give you idea about the quality of the image
```

```python
#flippint an image
fig= plt.figure(figsize=(10,10))
fig.add_subplot(1,2,1)
plt.imshow(np.flipud(img))
fig.add_subplot(1,2,2)
plt.imshow(np.fliplr(img))
```

```python
#Trimming an image
fig= plt.figure(figsize=(10,10))
fig.add_subplot(1,2,1)
plt.imshow(img)
plt.title('original')
img0=img[128:-128,128:-128,:]
fig.add_subplot(1,2,2)
plt.imshow(img0)
plt.title('Trimmed')
plt.show
```

```python
#padding black spaces

img=np.array(Image.open('Tiger_shutterstock.jpeg'))
img_grey = img.sum(2) / 255*3
img0=img_grey.copy()
img0=np.pad(img0,  ((100,100), (100,100)), mode = 'constant')
plt.imshow(img0)
```

```python
#negative image
fig = plt.figure(figsize=(10,10))
img_grey =(255*3)- img_grey
fig.add_subplot(1,2,1)
plt.imshow(img_grey)
plt.title('Negative of grey Image')
fig.add_subplot(1,2,2)
plt.imshow(img)
plt.title('Actula Image')
plt.show()
```
